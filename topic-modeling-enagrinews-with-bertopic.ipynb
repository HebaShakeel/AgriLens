```python
import pandas as pd
data = pd.read_csv('/kaggle/input/agriculture-data/cleaned_text_data.csv')
```

```python
print(data.columns)
```

```python
df = data.drop(columns=['file_name'])
```

```python
df.columns
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
df.head()
```

```python
df_filtered = df[(df['date'].dt.year >= 2020) & (df['date'].dt.year <= 2022)]
```

```python
texts = df_filtered['text'].tolist()
years = df_filtered['date'].dt.year.tolist()
```

```python
len(texts), len(years)
```

```python
from bertopic import BERTopic

# Initialize the topic model
topic_model = BERTopic(
    verbose=True,
    embedding_model="paraphrase-MiniLM-L12-v2",  # Compact and fast model
    min_topic_size=50  # Minimum size of a topic cluster
)

# Fit the model on your filtered texts
topics, _ = topic_model.fit_transform(texts)

# Get number of topics (including -1 outliers)
num_topics = len(topic_model.get_topic_info())

print("Number of topics found:", num_topics)
```

```python
topic_model.get_topic_info().head(25)
```

```python
topic_model.visualize_barchart(top_n_topics=9, height=700, topics=list(range(9)))
```

```python
topic_model.visualize_term_rank()
```

```python
topic_model.visualize_term_rank(log_scale=True)
```

```python
topic_model.visualize_topics(top_n_topics=50)
```

```python
topics_over_time = topic_model.topics_over_time(texts, topics, years)
```

```python
topic_model.visualize_topics_over_time(topics_over_time, top_n_topics=20, width=900, height=500)
```

```python
topics_per_class = topic_model.topics_per_class(texts, topics)
```

```python
topic_model.visualize_topics_per_class(topics_per_class, top_n_topics=10, width=900)
```
